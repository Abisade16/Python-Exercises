```python
# Array with NaN and infinite values
data = np.array([1, 2, np.nan, np.inf, -np.inf, 5])

# Check for NaN values
is_nan = np.isnan(data)
print("NaN values in the array:\n", is_nan)

# Check for infinite values
is_inf = np.isinf(data)
print("\nInfinite values in the array:\n", is_inf)
```

```python
# Replace NaN and infinite values with a specific number (e.g., 0)
cleaned_data = np.where(np.isnan(data) | np.isinf(data), 0, data)
print("\nArray after replacing NaN and inf values with 0:\n", cleaned_data)
```
