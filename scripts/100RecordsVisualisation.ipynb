```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
records= pd.read_csv('100 Records.csv')
records
```

```python
df= records.drop(columns=['E Mail','Date of Birth','Time of Birth', 'SSN', 'Password'])
df
```

```python
df.info()
```

```python
df[['Age in Yrs.', 'Age in Company(Years)']]=df[['Age in Yrs.', 'Age in Company (Years)']].astype(int)
df[['Age in Yrs.', 'Age in Company(Years)']]
```

```python
counts=df['Gender'].value_counts()
counts
```

```python
plt.pie(counts,labels=['Female','Male'], autopct='%1.1f%%')

plt.title('Percentage of Genders in the Company')
plt.legend()
plt.show()
```

```python
region=df['Region'].value_counts()
region
```

```python
plt.pie(region,labels=region.index, autopct='%1.1f%%')
plt.style.use('ggplot')

plt.title('Regions Occupied by Employees')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
gender_region=df[['Gender','Region']]
gender_region
```

```python
df1=gender_region.groupby('Gender')
df1
```

```python
females=df1.get_group('F')
males=df1.get_group('M')
female_count=females['Region'].value_counts()
male_count=males['Region'].value_counts()
```

```python
plt.bar(female_count.index,female_count.values, label='Females')
plt.bar(male_count.index,male_count.values, label='Males')

plt.title('Regions Occupied by each Gender')
plt.xlabel('Regions')
plt.ylabel('Number of Genders')
plt.legend()
plt.show
```

```python
gender=df.groupby('Gender')
```
