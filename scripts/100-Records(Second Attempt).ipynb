# Number of Employees by their Age Groups

18-24: Captures those just starting their careers, often in entry-level positions.

25-34: Represents those who are more established in their careers, possibly taking on more responsibility.

35-44: Captures mid-career professionals who are often moving into managerial or specialized roles.

45-54: Represents experienced professionals, often in senior roles or preparing for leadership.

55-64: Late career professionals, possibly nearing retirement or in high-level roles.

65+: While traditionally retirement age, this group can still include active professionals, especially in consulting or part-time roles.

```python
copy_records['Age in Yrs.'].values
```

```python
bins = [18, 25, 35, 45, 55, 65,100]
labels = ['25-34', '35-44', '45-54', '55-64', '18-24', '65+']

copy_records['Age Group'] = pd.cut(copy_records['Age in Yrs.'], bins=bins, labels=labels, right=False)
age_groups=list(copy_records['Age Group'].value_counts().values)



plt.pie(age_groups, labels=labels, shadow=True, startangle=90, autopct='%1.1f%%')
plt.title('Employee Age Groups')
plt.show()
```

From this, the age group with the highest number of employees is 25-34; which represents those who are more established in their careers, with decent experience in the fields they work.

The age group within the retirement age (65+) have no employees which seems reasonable.

Those within the entry level age (18-24) surprisingly have the least number of employees in the company. This isn't a disadvantage because the most employees in the company are young and active people.

## How much does each age group get averagely?

```python
grouped_data_by_age = copy_records.groupby('Age Group')['Salary'].mean()
grouped_data_by_age
```

```python
plt.bar(grouped_data_by_age.index,grouped_data_by_age.values)

plt.xlabel('Age Group')
plt.ylabel('Average Salary')
plt.title('Average Salary by Age Group')
plt.show()
```

Based on the bar chart, even though the entry level(18-24) are the least, they obtain the highest salary on average, compared to other age groups.

Followed by the late career professionals(55-64) due to their level of experience and the high roles they take in companies.

# Percentage of Female's Salary to that of Male

```python
gender_salary= copy_records.groupby('Gender')['Salary'].mean()
gender_salary
```

```python
average_salary = list(gender_salary.values)
labels= ['Female','Male']

plt.pie(average_salary,labels=labels, startangle=90, autopct='%1.1f%%')
plt.title("Gender's Average Salary in Percentage")
```

Male employees receive more salary on average than female employees 

# Who spent more years in the company; male or female?

```python
gender_AgeInCompany_average= copy_records.groupby('Gender')['Age in Company (Years)'].mean().round(2)
gender_AgeInCompany_average
```

```python
gender_AgeInCompany_sum= copy_records.groupby('Gender')['Age in Company (Years)'].sum()
gender_AgeInCompany_sum
```

```python
average_AgeInCompany = list(gender_AgeInCompany_average.values)
labels= ['Female','Male']
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.pie(average_AgeInCompany,labels=labels, autopct='%1.1f%%')
ax1.set_title('Average Years Spent in the Company by Genders')

ax2.bar(gender_AgeInCompany_sum.index,gender_AgeInCompany_sum.values)
ax2.set_xlabel('Gender')
ax2.set_ylabel('Sum of Years spent in the Company')
ax2.set_title('Sum of Years Spent in the Company by Genders')

plt.tight_layout()
plt.show()
```

On average, females spent more years in the company than males did.

Also, the total amount of years spent by females in the company is 50% more than males
