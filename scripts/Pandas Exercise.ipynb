# Data Selection and Indexing

```python
import pandas as pd

df = pd.read_csv('Documents/students.csv')

# Selecting the 'Name' column
name_column = df['Name']
print(name_column)
```

```python
df
```

```python
# Selecting 'Name' and 'City' columns
selected_columns = df[['Name', 'GPA']]
print(selected_columns)
```

```python
# Selecting the first row (position 0)
first_row = df.iloc[0]
print(first_row)
```

```python
# Selecting the first row and second column
first_row_second_col = df.iloc[0, 1]
print(first_row_second_col)
```

```python
# Selecting the first two rows and the first two columns
first_two_rows_cols = df.iloc[:2, :2]
print(first_two_rows_cols)
```

```python
# Selecting the row with index label 1
second_row = df.loc[1]
print(second_row)
```

```python
# Selecting the 'Name' and 'City' columns for row with index label 2
row2_name_city = df.loc[2, ['Name', 'GPA']]
print(row2_name_city)
```

```python
# Accessing the value at row label 1 and column 'City'
value = df.at[1, 'Name']
print(value)
```

```python
# Accessing the value at row position 1 and column position 2
value = df.iat[1, 2]
print(value)
```

```python
# Setting the 'Name' column as the index
df_with_index = df.set_index('Name')
print(df_with_index)
```

```python
df_with_index.loc['John Doe']
```

```python
# Resetting the index to default
df_reset = df_with_index.reset_index()
print(df_reset)
```

# Data Cleaning

```python
# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', None, 'Dave'],
    'Age': [24, None, 22, 19],
    'City': ['New York', 'Los Angeles', 'New York', None]
}
df = pd.DataFrame(data)

# Check for missing values
print(df.isnull())
```

```python
# Check for non-missing values
print(df.notnull())
```

```python
# Drop rows with missing values
df_cleaned = df.dropna()
print(df_cleaned)
```

```python
# Fill missing values with a default value, e.g., 0
df_filled = df.fillna(0)
print(df_filled)
```

```python
# Sample DataFrame with duplicates
data = {
    'Name': ['Alice', 'Bob', 'Alice', 'Dave'],
    'Age': [24, 24, 24, 19],
    'City': ['New York', 'Los Angeles', 'New York', 'Seattle']
}
df = pd.DataFrame(data)

# Drop duplicate rows
df_unique = df.drop_duplicates()
print(df_unique)
```

```python
# Replace 'New York' with 'NYC'
df_replaced = df.replace('New York', 'NYC')
print(df_replaced)
```

```python
# Rename columns
df_renamed = df.rename(columns={'Name': 'Full Name', 'City': 'Location'})
print(df_renamed)
```

```python
# Sample DataFrame with date strings
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Dave'],
    'JoinDate': ['2023-01-10', '2022-06-15', '2023-07-20', '2021-08-30']
}
df = pd.DataFrame(data)

# Convert JoinDate column to datetime format
df['JoinDate'] = pd.to_datetime(df['JoinDate'])
print(df)
```
