```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
hr = pd.read_csv('100 Records.csv')
hr
```

```python
hr.info()
```

```python
hr.dtypes['Age in Yrs.']
```

```python
hr[['Age in Yrs.', 'Age in Company(Years)']]=hr[['Age in Yrs.', 'Age in Company (Years)']].astype(int)
hr[['Age in Yrs.', 'Age in Company(Years)']]
```

```python
females = hr[hr['Gender'] == 'F']
print(females['Salary'].mean())
```

```python
males = hr[hr['Gender'] == 'M']
print(males['Salary'].mean())
```

On average, males get paid higher than females.

```python
genders = ['Male', 'Female']
salaries = [males['Salary'].sum(), females['Salary'].sum()]

plt.pie(salaries, labels=genders, autopct='%1.1f%%', startangle=90)
plt.title('Sum of Salaries by Gender')
plt.show()
```

Females receive more Salaries compared to Males

Even though the females got more than the males, the average says otherwise.
This maybe due to a lesser number of Females than males

```python
females = hr[hr['Gender'] == 'F']
print(females['Weight in Kgs.'].mean())
```

```python
males = hr[hr['Gender'] == 'M']
print(males['Weight in Kgs.'].mean())
```

```python
average = hr['Weight in Kgs.'].mean()
plt.plot(males['Weight in Kgs.'], label = 'Male')
plt.plot(females['Weight in Kgs.'], label = 'Female')
plt.axhline(y=average, color='r', linestyle='--', label='Average')

plt.xlabel('Index')
plt.ylabel('Weights')
plt.legend()
plt.show()
```

From the graph, majority of the females weigh below the average of weights while majority of the males weigh far above the average of weights.

Therefore, the Males weigh averagely more than females

```python
female_count = females['Region'].value_counts()
male_count = males['Region'].value_counts()
```

```python
regions_f = female_count.index
counts_f = female_count.values
regions_m = male_count.index
counts_m = male_count.values
plt.bar(regions_f, counts_f)
plt.bar(regions_m, counts_m)
plt.xlabel('Region')
plt.ylabel('Gender Counts')
plt.title('Number of Genders for each Region')

plt.show()
```
